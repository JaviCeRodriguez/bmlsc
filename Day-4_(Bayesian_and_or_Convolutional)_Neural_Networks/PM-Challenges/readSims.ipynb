```python
# Este código lee los datos simulados (código sims) y visualiza los primeros 20 eventos. 
# hecho por Carla Bonifazi (cbonifazi@unsam.edu.ar)
# 19 Oct 2023

import numpy as np
import matplotlib.pyplot as plt

# Leer el archivo1 para obtener la información de las simulaciones
info_sims = []
with open('simInput_100k.txt', 'r') as simInput:
    next(simInput)  # Saltar la primera línea (encabezado)
    for line in simInput:
        data = line.strip().split(',')
        info_sims.append(data)

# Leer el archivo2 para obtener los valores de energía de los eventos
pixValues = []
with open('simEvents_100k.txt', 'r') as simEvents:
    next(simEvents)  # Saltar la primera línea (encabezado)
    for line in simEvents:
        data = line.strip().split(',')
        pixValues.append(list(map(float, data[1:])))

# Número de eventos que deseas graficar 
N = 20

for event in range(N):
    # Obtener información del evento actual
    infoEvent = info_sims[event]

    # Redimensionar los valores de energía a una matriz de 8x8
    pixMatrix = np.array(pixValues[event]).reshape(8, 8)

    # Crear el gráfico 2D de los valores de energía
    plt.imshow(pixMatrix, cmap='viridis', origin='lower', extent=[-4, 4, -4, 4])
    plt.title(f"Evento {event + 1}:\n Sigma = {round(float(infoEvent[4]),3)}, Energy {round(float(infoEvent[5]),3)} e-")
    plt.xlabel('Posición en el eje X')
    plt.ylabel('Posición en el eje Y')
    plt.show()

```
