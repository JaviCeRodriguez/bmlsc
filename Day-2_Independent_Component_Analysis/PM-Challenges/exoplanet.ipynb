# ICA applied to HD189733

The pioneering method for detecting extrasolar planetas is the radial velocity (RV) technique.

In this method, the spectrum of a star is recorded on different orders in a *echelle* spectrograph, and a precise line-of-sight velocity is measured using the Doppler effect. An unseen planet induces variations in the line-of-sight velocity of the star, that can be detected by this method

Unfortunately, the star exhibits some intrinsic variability that can mimic the RV signal of an exoplanet, or hinder a precise determination of its physical parameters. This variability is called globally "stellar activity".

### ICA to the rescue!

Here we will use the velocities measured in the different spectral orders as microphone recordings in the cocktail party problem. We expect the signal of the planet to be present with equal amplitude in all orders, but the activity signal to vary significantly.

Download the data from the repository, which cotains a file for each visit (observation) of the star. In each file, you will find 39 measurements of the RV, one per order. You will also find a file providing mean values for each observation as well as the time of observation (`Log_HD189733.dat`).

### Reading the data

```python
import os
import pandas as pd

tbl = os.listdir('data')

# Read all table files
dfs = []
for tt in tbl:
    if tt.endswith('tbl'):
        a = pd.read_table('data/'+tt, sep='\t', skiprows=[1,], header=0)
        a = a.assign(filename = tt)
        dfs.append(a)

# Concatenate all values
df = pd.concat(dfs)
```
